```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
def generate_obstacle_time(lambda_val, size=10000):
    return np.random.exponential(1 / lambda_val, size)

lambda_values = [0.5, 5]
size = 10000

samples_lambda_0_5 = generate_obstacle_time(lambda_values[0], size)
samples_lambda_5 = generate_obstacle_time(lambda_values[1], size)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Histograma para lambda = 0.5
axes[0].hist(samples_lambda_0_5, bins=50, density=True, alpha=1, color='#FFC64D', label=r'$\lambda=0.5$', edgecolor='black')
axes[0].set_title(r'Histograma para $\lambda=0.5$')
axes[0].set_xlabel('Tempo entre obstáculos')
axes[0].set_ylabel('Densidade')

# Histograma para lambda = 5
axes[1].hist(samples_lambda_5, bins=50, density=True, alpha=1, color='#FFC64D', label=r'$\lambda=5$', edgecolor='black')
axes[1].set_title(r'Histograma para $\lambda=5$')
axes[1].set_xlabel('Tempo entre obstáculos')
axes[1].set_ylabel('Densidade')

plt.tight_layout()
plt.show()
```

```python
def generate_game_acceleration(elapsed_time, size=10000):
    mean = np.log(0.0225) + 0.00005 * elapsed_time
    sigma = 0.05 + 0.0001 * elapsed_time
    acceleration = np.random.lognormal(mean, sigma, size)
    return np.clip(acceleration, 0.015, 0.05)  # Limitar entre 0.015 e 0.05

elapsed_times = [0, 60, 300]  # 0s, 30s e 300s de jogo
size = 10000

samples_time_0 = generate_game_acceleration(elapsed_times[0], size)
samples_time_30 = generate_game_acceleration(elapsed_times[1], size)
samples_time_60 = generate_game_acceleration(elapsed_times[2], size)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, (samples, elapsed) in enumerate(zip([samples_time_0, samples_time_30, samples_time_60], elapsed_times)):
    axes[i].hist(samples, bins=50, density=True, alpha=1, color='#FFC64D', label=f'Tempo = {elapsed}s', edgecolor='black')
    axes[i].set_title(f'Histograma da Aceleração ({elapsed}s)')
    axes[i].set_xlabel('Aceleração')
    axes[i].set_ylabel('Densidade')
    axes[i].set_xlim(0.015, 0.05)

plt.tight_layout()
plt.show()
```

```python
def generate_cloud_time():
    """Generate the time for the next cloud."""
    return abs(random.gauss(5, 2))

samples = [generate_cloud_time() for _ in range(10000)]

plt.figure(figsize=(8, 5))
plt.hist(samples, bins=50, edgecolor='black', alpha=1, color="#FFC64D")
plt.xlabel("Tempo Gerado")
plt.ylabel("Frequência")
plt.title("Histograma da Função generate_cloud_time()")
plt.grid(False)
plt.show()
```

```python
def generate_cloud_height(height):
    """Returns a random vertical position for the cloud."""
    return random.randint(0, height // 2)

height = 300

samples_height = [generate_cloud_height(height) for _ in range(10000)]

plt.figure(figsize=(8, 5))
plt.hist(samples_height, bins=50, edgecolor='black', alpha=1, color="#FFC64D")
plt.xlabel("Altura Gerada")
plt.ylabel("Frequência")
plt.title("Histograma da Função generate_cloud_height(300)")
plt.grid(False)
plt.show()
```

```python
def generate_select_obstacle(obstacle_choices: list):
    """Select an obstacle based on specified probabilities."""
    return random.choices(obstacle_choices, weights=[0.6, 0.4], k=1)[0]

obstacle_choices = ["Small Tree", "Big Tree"]

samples_obstacles = [generate_select_obstacle(obstacle_choices) for _ in range(10000)]

plt.figure(figsize=(8, 5))
plt.hist(samples_obstacles, bins=range(3), 
         edgecolor='black', alpha=1, align='left', 
         rwidth=0.8, color="#FFC64D")

plt.xticks([0, 1], obstacle_choices)
plt.xlabel("Tipo de Árvore")
plt.ylabel("Frequência")
plt.title("Histograma da Função generate_select_obstacle([\"Small Tree\", \"Big Tree\"])")
plt.grid(False)
plt.show()
```

```python


def generate_num_trees(sprite_max_elements: int):
    """Get the number of trees to display."""
    return random.randint(1, sprite_max_elements)

sprite_max_elements = 4

samples_trees = [generate_num_trees(sprite_max_elements) for _ in range(10000)]

plt.figure(figsize=(8, 5))
plt.hist(samples_trees, bins=range(1, sprite_max_elements + 2), edgecolor='black', alpha=1, align='left', rwidth=0.8, color="#FFC64D")
plt.xticks(range(1, sprite_max_elements + 1))
plt.xlabel("Número de Árvores")
plt.ylabel("Frequência")
plt.title("Histograma da Função generate_num_trees(4)")
plt.grid(axis='y')
plt.grid(False)
plt.show()
```
